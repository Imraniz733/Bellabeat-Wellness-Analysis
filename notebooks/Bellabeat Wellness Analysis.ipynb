```python
# Summary of activity minutes
activity_minutes = daily_activity[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].mean()

# Bar plot to visualize the average minutes spent in each activity type
plt.figure(figsize=(10,6))
bars = activity_minutes.plot(kind='bar', color=['red', 'orange', 'green', 'grey'])

plt.title('Average Time Spent in Different Activity Levels', fontsize=16)
plt.ylabel('Minutes', fontsize=12)
plt.xlabel('Activity Type', fontsize=12)

plt.xticks(rotation=45, ha='right', fontsize=10)

for index, value in enumerate(activity_minutes):
    plt.text(index, value + 10, f'{int(value)} min', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('improved_average_time_spent.png')
plt.show()
```

#### Analysis
#### Very Active Minutes (Red Bar): 
##### Average time: 19 minutes. Represents low engagement in high-intensity exercises like running.

#### Fairly Active Minutes (Orange Bar): 
##### Average time: 13 minutes. Indicates low engagement in moderate activities such as brisk walking.

#### Lightly Active Minutes (Green Bar): 
##### Average time: 185 minutes. Shows moderate daily movement through casual walking and light tasks.

#### Sedentary Minutes (Gray Bar): 
##### Average time: 992 minutes (over 16 hours). Indicates that most of the day is spent in minimal physical activity.
##### The stark contrast between sedentary minutes and the other activity levels suggests that users are overwhelmingly inactive for most of the day. The engagement in light activity is far greater than in moderate or vigorous activities.

#### The total steps taken by users each day indicate a general trend of inactivity among the majority. To gain further insights, we will visualize the total step counts over time to examine these trends more closely.

```python
daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'])

daily_activity['Smoothed_Steps'] = daily_activity['TotalSteps'].rolling(window=7).mean()

sns.set(style='whitegrid')

plt.figure(figsize=(12, 6))

sns.lineplot(x='ActivityDate', y='TotalSteps', data=daily_activity, label='Total Steps', errorbar='sd', color='blue', alpha=0.6)

sns.lineplot(x='ActivityDate', y='Smoothed_Steps', data=daily_activity, label='Smoothed Steps (7-day avg)', color='red', linewidth=2)

plt.xticks(rotation=45, ha='right')

plt.title('Total Steps Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Steps', fontsize=12)

plt.ylim(0, daily_activity['TotalSteps'].max() + 1000)

plt.grid(True)

plt.tight_layout()
plt.savefig('total steps over time modified.png')

plt.show()
```

#### Analysis

#### High Variability: 
##### Daily step counts fluctuate from 20,000–30,000 steps to below 5,000 steps.

#### Stable Trend: 
##### The 7-day moving average indicates consistent activity levels despite daily spikes.

#### Activity Peaks: 
##### Temporary increases in steps suggest specific events or more intense physical activity.

#### Most daily step counts remain within the 5,000–10,000 range, with no significant long-term trend.
